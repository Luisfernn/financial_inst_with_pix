```python
path = '../data/pix_data.json'
import pandas as pd
import json
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
with open(path) as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head(10)    
```

```python
df.info()
```

```python
primeiras_palavras = df['nome_reduzido'].str.split().str[0].str.replace(r'[.,]', '', regex=True).str.upper()

perfil_siglas = primeiras_palavras.value_counts().reset_index()
perfil_siglas.columns = ['sigla_detectada', 'quantidade']

print(perfil_siglas)
```
