```python
import pandas as pd
import os

# --- EXPLICAÇÃO DO CAMINHO ---
# 1. '..' sai da pasta 'notebooks' e volta para a raiz
# 2. 'data' entra na pasta de dados
# 3. 'silver' entra na pasta silver
# 4. 'seu_arquivo.parquet' é o alvo
caminho_parquet = os.path.join('..', 'data', 'silver', 'final_institutions.parquet')

print(f"Buscando arquivo em: {os.path.abspath(caminho_parquet)}")

try:
    df = pd.read_parquet(caminho_parquet)
    
    # Pegamos tudo que tem menos de 5 caracteres para o diagnóstico
    rebeldes = df[df['categoria'].astype(str).str.len() < 5]['categoria'].unique()

    print("\n--- DIAGNÓSTICO DE CÉLULAS REBELDES ---")
    if len(rebeldes) == 0:
        print("Nenhum valor 'curto' ou vazio encontrado na coluna 'categoria'.")
    else:
        for val in rebeldes:
            print(f"Valor visual: '{val}'")
            print(f"Representação real (repr): {repr(val)}") 
            print(f"Tipo do dado: {type(val)}")
            print(f"Tamanho da string: {len(str(val))}")
            print("-" * 40)

    print(f"\nTotal de NaN (nulos reais): {df['categoria'].isna().sum()}")

except FileNotFoundError:
    print(f"\nERRO: Arquivo não encontrado!")
    print(f"O Python tentou buscar em: {os.getcwd()}")
    print("Verifique se o nome do arquivo .parquet está correto.")
```
